# **Task**

Create a joint plot using the seaborn visualization module. The data must be from the seaborn data repository and feel free to use any parameters.

```python
#Import libraries

import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#List all built-in datasets in the seaborn data repository

sns.get_dataset_names()
```

```python
#Load the 'Planets' dataset

planets = sns.load_dataset('planets')
```

```python
#Check the first 5 rows of the dataset

planets.head()
```

```python
#Check for the structure of the dataset

planets.info()
```

```python
#Check the summary statistics of the dataset

planets.describe()
```

# Visualization

```python
#Create a joint plot

sns.jointplot(
    data=planets,
    x="orbital_period",
    y="mass",
    kind="scatter",
    height=8
)

plt.title("Orbital Period vs Mass of Discovered Planets")
plt.show()
```
