# **Project: Student Grade Management System**

Project Description

The program allows users to:

1.Add student names and scores in three subjects (Math, Science, English).

2.Calculate the average score for each student.

3.Assign a grade based on the average score.

4.Display all student data (name, scores, average, and grade).

5.Use a menu-driven interface to interact with the program (terminal works)

```python
from statistics import mean
```

```python
#Use dictionaries to store students data with data types and input student details

student_data: dict[str, list[str | float]] = {
    'Name': ['Joseph', 'Tolani', 'Ruth', 'Stella', 'David'],  #Strings
    'Math': [40.0, 30.0, 25.0, 60.0, 55.0],  #Floats
    'Science': [70.0, 50.0, 80.0, 50.0, 30.0],  #Floats
    'English': [78.0, 90.0, 60.0, 50.0, 85.0]  #Floats
}

print("Student added successfully!\n")
```

```python
#Calculate Average Scores Using a Loop

average_scores = {}
for i in range(len(student_data['Name'])):
    name = student_data['Name'][i]
    avg_score = (student_data['Math'][i] + student_data['Science'][i] + student_data['English'][i]) / 3
    average_scores[name] = avg_score
```

```python
#Display result

for name, avg in average_scores.items():
    print(f"{name}: Average Score = {avg:.2f}")
```

```python
#Create a function and assign a grade based on the average score using conditional statement

def assign_grade(average_scores):
    if average_scores >= 60:
        return 'B'
    elif average_scores >= 55:
        return 'C'
    else:
        return 'D'
```

```python
##Display all student data (name, scores, average, and grade)

print("Student Records:\n")
print("Name       | Math  | Science | English | Average | Grade")
print("------------------------------------------------------")

for i in range(len(student_data['Name'])):
    name = student_data['Name'][i]
    math = student_data['Math'][i]
    science = student_data['Science'][i]
    english = student_data['English'][i]
    average = average_scores[name]
    grade = assign_grade(average)
    print(f"{name:<10} | {math:<5} | {science:<8} | {english:<8} | {average:<8.2f} | {grade}")
```
